```python
# 시계열 데이터로 데이터의 형태를 변경
# pd.to_datetime(변경할 데이터, format = "")
t = '23/04/05'
pd.to_datetime(t, format="%y/%m/%d")              # 대문자 : 4글자 / 소문자 : 2글자
```

```python
# puchase_data 컬럼을 시계열로 변경
uriage['purchase_date']
uriage['purchase_date'] = pd.to_datetime(uriage['purchase_date'], format = '%Y-%m-%d %H:%M')
```

```python
# 원본데이터를 생성
orgine_data = uriage.copy()

# 잘못 실행해서 변수가 바뀌었을때 원본 데이터로 다시 바꾸는 법
uriage = orgine_data.copy()
```

```python
# item_name 컬럼의 데이터를 모두 대문자로 변경
# upper() : 문자열을 모두 대문자 변환
# str.upper() 사용하여 문자열을 대문자로 변경하는 방식
uriage['item_name'] = uriage['item_name'].str.upper()
```

```python
# apply() 함수를 이용하여 글자를 모두 소문자로 변경
# apply()는 스리즈형태데이터.apply(함수) --> 스리즈의 데이터들을 하나씩 함수안에 넣어서 결과값으로 새로운 스리즈 형태의 데이터를 생성
# 새로운 스리즈형태의 데이터를 생성


def change(x) :
    result = x.lower()
    return result

uriage['item_name'] = uriage['item_name'].apply(change)
```

```python
# apply(lambda 함수)    def에서 함수 선언하고 호출하는 과정이 귀찮아서 선언과 동시에 호출로 간단하게 실행하는 함수.

uriage['item_name'] = uriage['item_name'].apply(lambda x : x.upper())
```

```python
# dataframe 형태에서 apply() 사용할 수 있다

def change(x) :
    print(x)
    return x

uriage.apply(change)
```

```python
# item_name에 있는 공백을 제거한다.
# 공백을 제거하는 함수 : strip() -> 문자열의 좌우 공백을 제거 -> 현재 데이터에서는 strip() 사용 불가
# 문자열 사이의 공백까지 제거하려면 사용할 함수의 이름?
# replace(기준문자열, 대체할문자열)
# replace(" ", "")
uriage["item_name"].str.replace(" ","").str.upper()
```

1. 상품A의 item_price의 값은 결측치 중 상품A이면 이 값을 치환
2. 나머지 상품들도 같은 작업을 반복

```python
# 상품A의 item_price을 출력
# 행의 조건? 결측치가 아니다, item_name이 상품A이다.
# 열의 조건? item_price

# 조건1) 결측치가 아닌 조건
flg_null = uriage['item_price'].isna()
# uriage.isna().any(1)  같은 코드지만, 시간이 더 걸린다.
```

```python
# 결측치인 값들을 출력
uriage.loc[flg_null]
```

```python
# item_name이 상품A인 경우만 출력
uriage.loc[uriage['item_name'] == '상품A']
```

```python
# 상품A이고 item_price가 결측치인 경우
uriage.loc[(flg_null) & (uriage['item_name'] == '상품A')]
```

```python
# 상품A이고 item_price가 결측치가 아닌 경우
uriage.loc[(uriage['item_name'] == '상품A') & ~(flg_null)]
```

```python
price = uriage.loc[(uriage['item_name'] == '상품A') & ~(flg_null), 'item_price'].value_counts().index[0]
```

```python
# 결측치에 값을 치환하는 함수
# fillna(n) : n의 값으로 결측치를 채운다. 
uriage.loc[(flg_null) & (uriage['item_name'] == '상품A')].fillna(price)
```

```python
uriage.loc[(flg_null) & (uriage['item_name'] == '상품A'), 'item_price'] = price
```

```python
# item_name의 리스트를 출력
name_list = uriage['item_name'].unique()
```

```python
for i in name_list :
    # item_price의 값을 구한다.
    # item_name이 i와 같고 결측치가 아닌 데이터에서 price를 출력
    price = uriage.loc[~(flg_null) & (uriage['item_name'] == i), 'item_price'].value_counts().index[0]

    # item_name이 i와 같고 결츨치인 데이터에서 item_price에 price라는 변수를 삽입
    uriage.loc[(flg_null) & (uriage['item_name'] == i), 'item_price'] = price
```
