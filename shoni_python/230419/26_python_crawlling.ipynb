```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
requests.get('https://finance.naver.com/marketindex/exchangeList.naver')
# 200이라는 뜻은, 정상적으로 접속했고, 데이터 가져왔다는 뜻
```

```python
# 실제 사이트에서 html 정보를 받아와서 변수에 저장
data = requests.get('https://finance.naver.com/marketindex/exchangeList.naver').text
```

```python
# BeatifulSoup을 이용해서 데이터를 파싱
soup = bs(data, 'html.parser')
```

```python
print(soup)
```

```python
# 하나밖에 없으니까 find로 찾는다
soup.find('tbody')
# find는 html안에서 해당 태그의 첫번째 정보를 출력
# 두번째 값을 찾으려면? - 일단 전체 가져와야 한다. find_all
# find_all() : 해당 태그의 전체 정보를 리스트의 형태로 출력
```

```python
# td나 tr로 쪼갠다

# td를 기준으로 출력
# soup.find_all('td')[0].get_text().replace('\n', '').strip('\t')
# find_all해서 첫번째값. 텍스트 가져오기. 줄바꿈,탭 제거

# find_all : 해당 태그의 전체 정보를 리스트의 형태로 출력
data_list = soup.find_all('td')
```

```python
# data_list를 이용하여 2차원 배열로 새로운 리스트를 생성
# 1차원 리스트 생성
# [[7개씩], [7개씩], [7개씩], [7개씩]] -> 2차원 리스트
_list = []
for i in data_list :
    x = i.get_text()
    x = x.replace("\n", '')
    x = x.strip('\t')
    _list.append(x)
_list
```

```python
# 리스트 2차원으로 나누기

# len(_list)/7        # float
len(_list) // 7       # int
```

```python
# list(range(0, (len(_list)/7), 1))    : float라 오류
list(range(0, (len(_list)//7), 1))   # : int라 결과 나온다.
```

```python
result = []
for i in range(0, (len(_list)//7), 1) :
    x = []
    for j in range(i*7, (i+1)*7, 1) :
        x.append(_list[j])
    result.append(x)
result
```

```python
# 0부터 7 전 까지
_list[0:7]
```

```python
# 위 과정을 더 간단하게 해보자
result = []
for i in range(len(_list)//7) :

    result.append(_list[i*7:(i+1)*7])
result
```

```python
# DB에 table 행성 (naver)
# 필드는 7개 모두 필드의 타입은 문자열 구성
# 통화명 필드는 primary key

sql = """
    create
    table
    `naver` (
        `통화명` varchar(16) primary key,
        `매매기준율` varchar(16),
        `현찰(구매)` varchar(16),
        `현찰(판매)` varchar(16),
        `송금(보낼때)` varchar(16),
        `송금(받을때)` varchar(16),
        `미화환산율` varchar(16)
    )
"""
db.sql_query(sql)
```

```python
# naver라는 이름으로 생성한 테이블에 데이터 삽입
# 데이터를 삽입하는 insert문을 작성

sql = """
    insert
    into
    `naver` 
    values (
        '미국 USD',
        '1,315.50',
        '1,338.52',
        '1,292.48',
        '1,328.30',
        '1,302.70',
        '1.000'
    )
    
"""
db.sql_query(sql)
```
