```python
df = pd.read_csv("../csv/AMZN.csv", index_col= 'Date')
```

```python
import numpy as np
```

```python
from datetime import datetime
```

```python
# 결측치, 이상치 제거
df = df.loc[~df.isin([np.nan, np.inf, -np.inf]).any(1)]
```

```python
df.index = pd.to_datetime(df.index, format = '%Y-%m-%d')
```

```python
# 새로운 파생변수 년-월의 데이터를 대입
df['STD_YM'] = list(map(lambda x : datetime.strftime(x, '%Y-%m'), df.index))
```

```python
df.head()
```

```python
# 월별 마지막 인덱스 추출
month_list_df = pd.DataFrame()

_list = df['STD_YM'].unique()
for i in _list :
    last_df = df.loc[df['STD_YM'] == i].tail(1)
    month_list_df = pd.concat([month_list_df, last_df])
month_list_df.head(2)
```

```python
month_list_df['STD_YM'] = pd.to_datetime(month_list_df['STD_YM'], format = '%Y-%m-%d')
```

```python
nov_df = month_list_df[month_list_df['STD_YM'].dt.month == 11]
nov_df
    
```

```python
# 월별 첫번째 인덱스 추출
month_list_df2 = pd.DataFrame()

_list = df['STD_YM'].unique()
for i in _list :
    first_df = df.loc[df['STD_YM'] == i].head(1)
    month_list_df2 = pd.concat([month_list_df2, first_df])
month_list_df2.head()
```

```python
month_list_df2['STD_YM'] = pd.to_datetime(month_list_df2['STD_YM'], format = '%Y-%m-%d')
```

```python
apr_df = month_list_df2[month_list_df2['STD_YM'].dt.month == 4]
apr_df
```

```python
df1 = pd.concat([nov_df, apr_df]).sort_values('Date')
```

```python
df1['BF_6M'] = df1['Open'].shift(1).fillna(0)
```

```python
df1.head()
```

```python
# 수익률 계산

# 파생변수 생성
df1['retrun'] = 1
```

```python
df1.head(2)
```

```python
for i in df1.index :
    if df1.loc[i,'STD_YM'].isin(4) :
        df1['return'] += df.loc[i,'Close'] - df.loc[i, 'BF_6M'] / df.loc[i, 'BF_6M']
```

```python
# 11월 첫날
df.loc['2000-11'].iloc[0]['Open']
```

```python
for i in range(2000, 2020, 1):
    buy_mon = str(i) + "-11"
    sell_mon = str(i+1) + "-04"
    print(buy_mon, sell_mon)
```

```python
acc_rtn = 1
for i in range(2000, 2019, 1) :
    buy_mon = str(i) + "-11"
    sell_mon = str(i+1) + '-04'

    # 구입가 저장
    buy = df.loc[buy_mon].iloc[0]['Open']
    sell = df.loc[sell_mon].iloc[-1]['Close']

    # 수익률
    rtn = (sell - buy) / buy + 1
    
    # 누적 수익률
    acc_rtn *= rtn
```

```python
acc_rtn
```

```python
# buyandhold 인 경우
buy = df.loc['2000-11'].iloc[0]['Open']
sell = df.loc['2019-04'].iloc[-1]['Close']

rtn = (sell-buy)/buy + 1

rtn
```

```python
# CAGR(연복리 수익률)
# ((수익률) ** (1/투자기간)) - 1

# 할로윈 CAGR
CAGR = (acc_rtn ** (1/20)) - 1
print(CAGR * 100)
```

```python
# 바이앤홀드 CAGR
CAGR = (rtn ** (1/20)) - 1
print(CAGR * 100)
```

```python
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
start = datetime(year = 2000, month = 11, day = 1)
start
```

```python
end = start + relativedelta(months = 5) # 5월 초가 아니라, 4월 말 종가 필요하므로 5개월 뒤로
end
```

```python
print(start.strftime("%Y-%m"), end.strftime("%Y-%m"))
```

```python
def six_month(df, start = 2000, end = 2019, month = 11):
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'], format = '%Y-%m-%d')
        df.set_index('Date', inplace = True)

    # 누적수익률 변수를 생성
    acc_rtn = 1

    for i in range(start, end) :
        start_ = datetime(year = i, month = month, day = 1)
        end_ = start_ + relativedelta(months = 5)

        buy_mon = start_.strftime("%Y-%m")
        sell_mon = end_.strftime("%Y-%m")

        buy = df.loc[buy_mon].iloc[0]['Open']
        sell = df.loc[sell_mon].iloc[-1]['Close']

        rtn = (sell - buy) / buy + 1
        
        acc_rtn *= rtn

    return acc_rtn
```

```python
data = pd.read_csv("../csv/AMZN.csv")
```

```python
six_month(data)
```

```python
for i in range(1, 13, 1):
    print(f"{i}월 :", six_month(data, month = i))
```
