```python
# 인덱스를 시계열 데이터로 변경
df.index = pd.to_datetime(df.index, format = '%Y-%m-%d')
```

```python
df['STD_YM'] = list(map(lambda x : datetime.strftime(x, '%Y-%m'), df.index))
```

```python
month_list_df = pd.DataFrame()

_list = df['STD_YM'].unique()
for i in _list :
    last_df = df.loc[df['STD_YM'] == i].tail(1)
    month_list_df = pd.concat([month_list_df, last_df])

month_list_df
```

```python
# 전월 종가 파생변수 생성
month_list_df['BF_1M'] = month_list_df['Adj Close'].shift(1).fillna(0)['005930.KS']
month_list_df
```

```python
# 수익률 계산

# 파생변수 생성
df['return'] = 1

# 초기값
rtn = 1.0
buy = 0.0
sell = 0.0

for i in df.index :             # .index를 사용할 수 있는 이유는, index가 중복되는 값이 없이 다 unique한 값이기 때문이다.
    
    # 구매한 날을 체크, 구매가 지정
    if (df.loc[i, 'trade'] == 'buy') & (df.shift(1).loc[i, 'trade'] == '' ) :
        buy = df.loc[i, 'Adj Close']
        print('진입일 :', i, '구입 가격 :', buy)
    
    # 판매한 날을 체크, 판매가 지정, 수익률 계산    
    elif (df.loc[i, 'trade'] == '') & (df.shift(1).loc[i, 'trade'] == 'buy') :
        sell = df.loc[i, 'Adj Close']
        rtn = (sell - buy) / buy + 1
        df.loc[i, 'return'] = rtn
        print('판매일 :', i, '판매 가격 ;', sell, '수익률 :', rtn)
```
